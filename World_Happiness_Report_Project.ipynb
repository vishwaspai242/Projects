# Removing Outliers:

Z-Score Technique

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df2))
```

```python
# removing outliers
df_new=df2[(z<3).all(axis=1)]
print("Old Dataframe",df2.shape)
print("New Dataframe",df_new.shape)
print("total_dropped_rows",df2.shape[0] - df_new.shape[0])
```

# Since the y (Target) has continuous values we will use Linear Regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
lm = LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=45)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
lm.coef_
```

```python
predlm=lm.predict(x_test)
predlm
```

```python
mean_squared_error(y_test,predlm)
```

```python
np.sqrt(mean_squared_error(y_test,predlm))
```

```python
print(r2_score(y_test,predlm))
```

# L1 -- Lasso Regression, L2 -- Ridge Regression and Elastic Net

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
ls=Lasso(alpha=.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
predlasso=ls.predict(x_test)
print(mean_squared_error(y_test,predlasso))
print('Mean squared error:',np.sqrt(mean_squared_error(y_test,predlasso)))
```

```python
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
print(rd.score(x_train,y_train))
predridge=rd.predict(x_test)
print(mean_squared_error(y_test,predridge))
print('Mean squared error:',np.sqrt(mean_squared_error(y_test,predridge)))
```

```python
enr = ElasticNet(alpha=0.0001)
enr.fit(x_train,y_train)
print(enr.score(x_train,y_train))
enrpred=enr.predict(x_test)
print(mean_squared_error(y_test,enrpred))
print('Mean squared error:',np.sqrt(mean_squared_error(y_test,enrpred)))
```
